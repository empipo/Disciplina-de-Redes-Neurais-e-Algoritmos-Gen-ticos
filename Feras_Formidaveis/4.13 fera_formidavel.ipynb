### Função para criar candidato:

Vamos tratar esse problema como o problema da mochila e como foi resolvido na videoaula do professor. Dessa forma utilizaremos caixas binárias com algumas modificações. Primeiramente, será criado uma função para gerar valores aleatórios de 3 elementos (x, y, z) que respeitem as seguintes regras:

$$x + y + z = 100$$
$$x \geq 5$$
$$y \geq 5$$
$$z \geq 5$$

Um candidato nesse problema será uma lista de elementos iguais a 0 ou diferentes de 0, na qual, apenas 3 itens da lista tem valor diferente de 0 de forma que respeitem as regras propostas. Para isso, será criado uma função que escolhe 3 possíveis número $x, y \text{ e } z$ que estejam de acordo com as regras. E posteriormente uma função que cria um candidato válido baseado nessa função.

```python
def gerar_valores_100g():
    '''
    Args: Nenhum
    
    Return: Gerá 3 números x, y, z que seguem as regras:
    x + y + z = 100
    x >= 5
    y >= 5
    z >= 5
    '''
    while True:
        x = round(random.uniform(5, 90), 2)
        y = round(random.uniform(5, 100 - x - 5), 2)
        z = round(100 - x - y, 2)

        if z >= 5:
            valores = [x, y, z]
            random.shuffle(valores)
            return valores
        
```

```python
v = gerar_valores_100g()
print(f"Valores aleatórios: {v}")
print(f"A soma dos valores aleatórios é: {sum(v)}")
```

### Função que cria candidato:

```python
def cria_candidato_binario_ligas():
    '''
    Args: Nenhum
    
    Return: Gera um candidato que é uma lista de 92 elementos, com 3 elementos preenchidos com os números escolhidos 
    pela função gerar_valores_100g()
    
    '''
    candidato = [0] * NUM_ITENS_TOTAIS

    posicoes = random.sample(range(NUM_ITENS_TOTAIS), 3)

    valores = gerar_valores_100g()
    # Atribuir os valores nas posições sorteadas
    for pos, val in zip(posicoes, valores):
        candidato[pos] = val

    return candidato
    
```

### Função para criar população:

```python
def cria_pop_binario_ligas(tamanho_populacao = TAMANHO_POPULACAO):
    '''
    Args: 
    - tamanho_populacao = quantidade de candidatos que o usuário quer fazer
    
    Return:
    - retorna uma população de candidatos da função cria_candidato_binario_ligas()
    '''
    populacao = []
    
    for _ in range(tamanho_populacao):
        populacao.append(cria_candidato_binario_ligas())

    return populacao
```

### Função para calcular os parâmetros da liga:

Uma função que vai retornar o preço total da liga metálica, os 3 elementos que formarão essa liga e as gramas utilizadas por cada elemento correspondente para alcançar esse custo.

```python
def calcula_liga(candidato, preco, ordem_dos_itens):
    """
    Args:
      candidato: lista representando quais itens estão na liga e com quantas gramas cada um estpa
      itens: dicionário com os preços de cada elemento
      ordem_dos_itens: ordem dos itens da lista `candidato`

    Return: uma tupla de 3 itens
    - custo da liga
    - lista de elementos que será usado na liga
    - lista da quantidade de gramas de cada elementos que serão usados na liga
    """
    custo = 0
    elementos= []
    gramas_por_elemento = []
    
    for quantidade, nome_item in zip(candidato, ordem_dos_itens):
        if quantidade != 0:
            custo += preco[nome_item] * (quantidade/1000)
            elementos.append(nome_item)
            gramas_por_elemento.append(quantidade)
    return custo, elementos, gramas_por_elemento
```

### Função para calcular a função objetivo para um candidato

```python
def funcao_objetivo_liga(candidato, preco, ordem_dos_itens):
    """Computa a funcao objetivo de um candidato no problema da liga

    Args:
      candidato: lista representando quais itens estão na mochila
      itens: dicionário com os pesos e valores dos itens
      ordem_dos_itens: ordem dos itens da lista `candidato`
    """
    # pega somente o preço da liga metálica
    fitness = calcula_liga(candidato, preco, ordem_dos_itens)[0]
    
    return fitness
```

### Função para calcular a função objetivo para uma população:

```python
def funcao_objetivo_liga_pop(populacao, preco, ordem_dos_itens):
    fitness = []

    for individuo in populacao:
        fitness.append(
            funcao_objetivo_liga(individuo, preco, ordem_dos_itens)
        )

    return fitness
```

### Função de seleção:

É um problema de maximização, dessa forma, vai ser utilizado a função de roleta de maximização.

```python
def funcao_selecao_roleta_max(populacao, fitness):
    """Realiza seleção da população pela roleta

    Args:
      populacao: lista contendo os individuos do problema
      fitness: lista contendo os valores computados da funcao objetivo

    """
    selecionados = random.choices(populacao, fitness, k=len(populacao))
    return selecionados
```

### Função de cruzamento:

A função de cruzamento tem como objetivo gerar dois filhos válidos a partir de dois pais, onde cada indivíduo representa uma liga composta por exatamente três elementos químicos diferentes. Cada elemento presente possui uma quantidade em gramas maior ou igual a 5, e a soma total das massas dos três elementos é exatamente 100 gramas. Os outros elementos (dos 92 totais) possuem valor zero. Para realizar o cruzamento, primeiro são identificados os índices dos elementos que têm valor diferente de zero em cada um dos pais, ou seja, os três elementos que fazem parte da liga de cada um. Em seguida, o cruzamento é feito da seguinte forma: o primeiro filho herda os mesmos índices de elementos do pai (ou seja, os mesmos três elementos químicos do pai), mas as quantidades nesses índices são copiadas da mãe. Já o segundo filho faz o oposto — ele herda os índices de elementos da mãe, mas os valores (quantidades em gramas) são os do pai. Com isso, cada filho possui exatamente três elementos diferentes com os mesmos critérios de validade: todos os elementos têm massa ≥ 5 g e a soma total continua sendo 100 g. Esse método garante que todos os indivíduos gerados durante o cruzamento permaneçam válidos para o problema.

```python
def cruzamento_ligas(pai, mae, chance_de_cruzamento):
    """
    Args:
      pai: lista representando um individuo
      mae: lista representando um individuo
      chance_de_cruzamento: float entre 0 e 1 representando a chance de cruzamento

    """
    if random.random() < chance_de_cruzamento:
      # Encontra índices dos elementos com valor > 0
      index_pai = [i for i, val in enumerate(pai) if val > 0]
      valores_pai = [val for val in pai if val > 0]

      index_mae = [i for i, val in enumerate(mae) if val > 0]
      valores_mae = [val for val in mae if val > 0]
      
      filho1 = [0] * NUM_ITENS_TOTAIS
      filho2 = [0] * NUM_ITENS_TOTAIS
      for i, val in zip(index_pai, valores_mae):
        filho1[i] = val
      for i, val in zip(index_mae, valores_pai):
        filho2[i] = val
          
      return filho1, filho2
        
    else:
        return pai, mae
```

### Função de mutação:

Vamos criar uma função de mutação que pode mutar qualquer gene do indivíduo. Se mutar um gene igual a um gene que já seja um número diferete de zero, os outros 2 elementos também serão mutados para que se mantenha a função $x + y + z = 100$. No entanto, se o gene mutado for diferente dos 3 elementos já selecionados do candidato, um dos outros 3 será retirado aleatoriamente e os outros 2 serão mutados. De forma geral, se um candidato for mutado nesses casos, todas as informações da quantidade de gramas seria alterada para manter as condições dos candidatos para eles continuarem válidos. O que se procura alterar com essa mutação, é o elemento utilizado no problema, não se preocupando em manter ou não a informação da quantidade de gramas armazenada em cada candidato.

```python
import random

def funcao_mutacao(populacao, chance_de_mutacao):
    """Realiza mutação 

    Args:
      populacao: lista contendo os indivíduos do problema
      chance_de_mutacao: float entre 0 e 1 representando a chance de mutação

    """
    
    for individuo in populacao:
        if random.random() < chance_de_mutacao:
            
            # Encontra índices dos elementos com valor > 0
            index_individuo = [i for i, val in enumerate(individuo) if val > 0]

            # Sorteia um gene para mutação
            gene = random.randint(0, len(individuo) - 1)
            valor_gene = individuo[gene]

            if valor_gene > 0:
                # Caso o gene já esteja entre os 3 usados, mutar todos os 3
                valores = gerar_valores_100g()

                for pos, val in zip(index_individuo, valores):
                    individuo[pos] = val

            else:
                # Gene novo, remove 1 dos 3, adiciona esse, e muta os 3
                removido = random.choice(index_individuo)
                novos_indices = [i for i in index_individuo if i != removido] + [gene]

                valores = gerar_valores_100g()

                individuo[removido] = 0.0
                for pos, val in zip(novos_indices, valores):
                    individuo[pos] = val

    return populacao
```

```python
hall_da_fama = []

for n in range(NUM_GERACOES):
    
    # Seleção
    fitness = funcao_objetivo_liga_pop(populacao, preco, ORDEM_DOS_ITENS)        
    selecionados = funcao_selecao_roleta_max(populacao, fitness)
    
    # Cruzamento
    proxima_geracao = []
    for pai, mae in zip(selecionados[::2], selecionados[1::2]):
        individuo1, individuo2 = cruzamento_ligas(pai, mae, CHANCE_DE_CRUZAMENTO)
        proxima_geracao.append(individuo1)
        proxima_geracao.append(individuo2)
    
    # Mutação
    funcao_mutacao(proxima_geracao, CHANCE_DE_MUTACAO)
    funcao_mutacao(proxima_geracao, CHANCE_DE_MUTACAO)
    
    # Atualização do hall da fama
    fitness = funcao_objetivo_liga_pop(proxima_geracao, preco, ORDEM_DOS_ITENS)
        
    maior_fitness = max(fitness)
    indice = fitness.index(maior_fitness)
    hall_da_fama.append(proxima_geracao[indice])    
    
    # Encerramento
    populacao = proxima_geracao
```
