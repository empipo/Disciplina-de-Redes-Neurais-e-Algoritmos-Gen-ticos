#### Função para criar candidato:

Será usado a mesma função da criada na fera formidável 4.13 para criar candidatos válidos. Os candidatos nesse problema serão listas de tamanho 92. Para cada item representar a quantidade de gramas de cada elemento, e 3 itens serão sorteados para terem quantidades de gramas diferentes de zero, mantendo as regras de candidatos do problema em que a soma das gramas dos elementos deve ser igual a 100 e todos os elementos devem ter uma quantidade mínima de 5 elementos. Mas a função abaixo apenas cria uma lista de 3 elementos que proporcionará as quantidades corretas.

```python
def gerar_valores_100g():
    '''
    Args: Nenhum
    
    Return: Gerá 3 números x, y, z que seguem as regras:
    x + y + z = 100
    x >= 5
    y >= 5
    z >= 5
    '''
    while True:
        x = round(random.uniform(5, 90), 2)
        y = round(random.uniform(5, 100 - x - 5), 2)
        z = round(100 - x - y, 2)

        if z >= 5:
            valores = [x, y, z]
            random.shuffle(valores)
            return valores
```

#### Função que cria candidato:

Agora, será feito uma função que cria um candidato válido baseado na função 'gerar_valores_100g()'

```python
def cria_candidato_ligas():
    '''
    Args: Nenhum
    
    Return: Gera um candidato que é uma lista de 92 elementos, com 3 elementos preenchidos com os números escolhidos 
    pela função gerar_valores_100g()
    '''
    
    #Cria lista
    candidato = [0] * NUM_ITENS_TOTAIS_PRECO

    #Sortea 3 posições da lista
    posicoes = random.sample(range(NUM_ITENS_TOTAIS_PRECO), 3)

    valores = gerar_valores_100g()
    # Atribuir os valores nas posições sorteadas
    for pos, val in zip(posicoes, valores):
        candidato[pos] = val

    return candidato
```

#### Função para calcular os parâmetros da liga:

Essa função foi alterada da função da fera 4.13, pois para o presente problema é importante retornar os custo, elementos e gramas utilizadas por elementos, mas além dessas parâmetros, é importante retornar 

```python
def calcula_liga(candidato, preco, peso_atomico, ordem_dos_itens):
    """
    Args:
      candidato: lista representando quais itens estão na liga e com quantas gramas cada um estpa
      preco: dicionário com os preços de cada elemento
      preco_atomico: dicionário com os pesos atomicos de cada elemento
      ordem_dos_itens: ordem dos itens da lista `candidato`

    Return: uma tupla de 3 itens
    - custo da liga
    - lista de elementos que será usado na liga
    - lista da quantidade de gramas de cada elementos que serão usados na liga
    """
    custo = 0
    elementos= []
    gramas_por_elemento = []
    soma_peso_atomico = 0
    
    for quantidade, nome_item in zip(candidato, ordem_dos_itens):
        if quantidade != 0:
            #calcular o custo
            custo += preco[nome_item] * (quantidade/1000)
            #verficar elemento
            elementos.append(nome_item)
            #quantidade de gramas
            gramas_por_elemento.append(quantidade)
            #soma dos pesos atômicos:
            soma_peso_atomico += peso_atomico[nome_item]
            
    return custo, elementos, gramas_por_elemento, soma_peso_atomico
```

### Função objetivo

Agora, é possível usar como fitness do algoritmo genetico como a função do fitness invertido para o parâmetro de custo que queremos maximizar, assim transformamos num problema de minimização tal qual o parâmetro de peso atômico: que é de minimização

$$\text{fitness invertido para o custo} = \dfrac{1}{1+f} \text{, onde f é o fitness do custo}$$

E o fitness nesse caso vai ser simplesmente o retorno do valor do custo 

Depois somamos

$$\text{fitness total} = \text{fitness invertido para o custo} + \text{fitnesse para o peso}$$

E buscamos o minimo desse fitness

```python
def funcao_custo_liga_leve(candidato, preco, peso_atomico, ordem_dos_itens):
    """Computa o custo de um candidato no problema da liga leve mais cara somente para o custo

    Args:
      candidato: lista representando quais itens estão na mochila
      preco: dicionário com os preços 
      peso: dicionário com os pesos
      ordem_dos_itens: ordem dos itens da lista `candidato`
    """
    # pega somente o preço da liga metálica
    custo = calcula_liga(candidato, preco, peso_atomico, ordem_dos_itens)[0]
    
    return custo
```

```python
def funcao_custo_liga_leve_pop(populacao, preco, peso_atomico, ordem_dos_itens):
    custo = []

    for individuo in populacao:
        custo.append(
            funcao_custo_liga_leve(individuo, preco, peso_atomico, ordem_dos_itens)
        )

    return custo
```

```python
def fitness_custo_liga_leve_custo(populacao, preco, peso_atomico, ordem_dos_itens):
    custo = funcao_custo_liga_leve_pop(populacao, preco, peso_atomico, ordem_dos_itens)
    
    fitness_invertido = [1 / (1 + c) for c in custo]
    
    return fitness_invertido
```

Agora calcular o fitness do peso atômico.

```python
def funcao_objetivo_liga_peso(candidato, preco, peso_atomico, ordem_dos_itens):
    """Computa a funcao objetivo de um candidato no problema da liga

    Args:
      candidato: lista representando quais itens estão na mochila
      itens: dicionário com os pesos e valores dos itens
      ordem_dos_itens: ordem dos itens da lista `candidato`
    """
    # pega somente o preço da liga metálica
    fitness = calcula_liga(candidato, preco, peso_atomico, ordem_dos_itens)[3]
    
    return fitness
```

```python
def fitness_peso_liga_leve(populacao, preco, peso_atomico, ordem_dos_itens):
    fitness = []

    for individuo in populacao:
        fitness.append(
            funcao_objetivo_liga_peso(individuo, preco, peso_atomico, ordem_dos_itens)
        )

    return fitness
```

E agora o fitness total para o problema de minimização:

```python
def fitness_total_pop(populacao, preco, peso_atomico, ordem_dos_itens):
    fitness_total = []
    
    fitness_custo = fitness_custo_liga_leve_custo(populacao, preco, peso_atomico, ordem_dos_itens)
    
    fitness_peso = fitness_peso_liga_leve(populacao, preco, peso_atomico, ordem_dos_itens)
    
    fitness_total = [num1 + num2 for num1, num2 in zip(fitness_custo, fitness_peso)]
    
    return fitness_total
```

#### Função de seleção: 



```python
def selecao_torneio_min(populacao, fitness, tamanho_torneio):
    """Faz a seleção de uma população usando torneio.

    Nota: da forma que está implementada, só funciona em problemas de
    minimização.

    Args:
      populacao: lista contendo os individuos do problema
      fitness: lista contendo os valores computados da funcao objetivo
      tamanho_torneio: quantidade de invíduos que batalham entre si retirados da população

    """
    selecionados = []

    for _ in range(len(populacao)):
        sorteados = random.sample(populacao, tamanho_torneio)

        fitness_sorteados = []
        for individuo in sorteados:
            indice_individuo = populacao.index(individuo)
            fitness_sorteados.append(fitness[indice_individuo])

        min_fitness = min(fitness_sorteados)
        indice_min_fitness = fitness_sorteados.index(min_fitness)
        individuo_selecionado = sorteados[indice_min_fitness]

        selecionados.append(individuo_selecionado)

    return selecionados
```

#### Função de cruzamento: 


```python
def cruzamento_ligas(pai, mae, chance_de_cruzamento):
    """
    Args:
      pai: lista representando um individuo
      mae: lista representando um individuo
      chance_de_cruzamento: float entre 0 e 1 representando a chance de cruzamento

    """
    if random.random() < chance_de_cruzamento:
      # Encontra índices dos elementos com valor > 0
      index_pai = [i for i, val in enumerate(pai) if val > 0]
      valores_pai = [val for val in pai if val > 0]

      index_mae = [i for i, val in enumerate(mae) if val > 0]
      valores_mae = [val for val in mae if val > 0]
      
      filho1 = [0] * NUM_ITENS_TOTAIS
      filho2 = [0] * NUM_ITENS_TOTAIS
      for i, val in zip(index_pai, valores_mae):
        filho1[i] = val
      for i, val in zip(index_mae, valores_pai):
        filho2[i] = val
          
      return filho1, filho2
        
    else:
        return pai, mae
```

#### Função de mutação: 



```python
import random

def funcao_mutacao(populacao, chance_de_mutacao):
    """Realiza mutação 

    Args:
      populacao: lista contendo os indivíduos do problema
      chance_de_mutacao: float entre 0 e 1 representando a chance de mutação

    """
    
    for individuo in populacao:
        if random.random() < chance_de_mutacao:
            
            # Encontra índices dos elementos com valor > 0
            index_individuo = [i for i, val in enumerate(individuo) if val > 0]

            # Sorteia um gene para mutação
            gene = random.randint(0, len(individuo) - 1)
            valor_gene = individuo[gene]

            if valor_gene > 0:
                # Caso o gene já esteja entre os 3 usados, mutar todos os 3
                valores = gerar_valores_100g()

                for pos, val in zip(index_individuo, valores):
                    individuo[pos] = val

            else:
                # Gene novo, remove 1 dos 3, adiciona esse, e muta os 3
                removido = random.choice(index_individuo)
                novos_indices = [i for i in index_individuo if i != removido] + [gene]

                valores = gerar_valores_100g()

                individuo[removido] = 0.0
                for pos, val in zip(novos_indices, valores):
                    individuo[pos] = val

    return populacao
```

### Utilizando a fronteira de Pareto:

Para esse problema multiobjetivo, será utilizado a fronteira de Pareto[1], os números que estão na fronteira é o conjunto de pontos onde um objetivo não pode ser melhorado sem prejuízo de outro. Esses pontos também são chamados de "não dominados". Enquanto isso, os pontos que não estão na fronteira, são os "dominados", pontos que podem melhorar um ou mais objetivo sem projuizo de outros pontos, dessa forma, são pontos que não são os pontos mínimos ou máximo de todos os objetivos.


O ponto C não está na fronteira de Pareto porque é dominado pelos pontos A e B. Os pontos A e B não são estritamente dominados por nenhum outro e, portanto, estão na fronteira. [2]

#### Função objetivo:

Para o presente problema, será realizado uma função objetivo bem diferente das habituais realizadas em aula, será usado a frente de pareto. Para isso será criado uma função que dirá se entre dois candidatos A e B: A é maior ou igual que B em todos os critérios que queremos maximizar ou menor ou igual em todos que queremos minimizar — assim dizemos que A domina B, e dessa forma, A é um elemento não dominado e está na fronteira de pareto. 

Queremos maximizar o custo e queremos minimizar o peso atomico, dessa forma, a função definida abaixo funciona da seguinte forma, se o candidato 1 tem custo maior ou igual ao candidato 2 e ao mesmo tempo tem peso atomico menor que o candidato 2, então o candidato 1 domina 2.

```python
def domina(candidato1, candidato2):
    '''
    Args:
    - candidato 1 e 2 são listas
    
    Return:
    Verifica se candidato 1 domina o candidato 2
    '''
    domina_1_2 = False
    parametros1 = calcula_liga(candidato1, preco, peso_atomico, ORDEM_DOS_ITENS_PESO)
    parametros2 = calcula_liga(candidato2, preco, peso_atomico, ORDEM_DOS_ITENS_PESO)
    if (parametros1[0] >= parametros2[0] and parametros1[3] <= parametros2[3]):
        domina_1_2 = True
    # candidato 1 domina 2
    return domina_1_2
```

Dessa forma, um indivíduo estará na fronteira de pareto se comparado com todos os demais elementos, ele tiver maior ou igual custo, e menor ou igual peso atomico. Isto é, ele não é dominado por ninguém. A nossa função objetivo irá funciona da seguinte forma: um indivíduo com fitness 0 é dito não dominado, isto é, ninguém domina ele, cada vez que um indivíduo for dominado terá um acréscimo de 1 ponto quando ele for dominado

```python
def funcao_objetivo_pareto(populacao, preco, peso_atomico):
    fitness_lista_pareto = [0 for _ in range(len(populacao))]
    
    for i in range(len(populacao)):
        contador_de_i = 0
        for j in range(len(populacao)):
            # se forem indivíduos iguais compare
            if i != j:
                if domina(populacao[j], populacao[i]):
                    contador_de_i += 1
                    
        fitness_lista_pareto[i] = contador_de_i
    return fitness_lista_pareto
                
```

Vamos verificar seu funcionamento

```python
populacao_teste_pareto = cria_pop_ligas(TAMANHO_POPULACAO)
print(funcao_objetivo_pareto(populacao_teste_pareto, preco, peso_atomico))
```

Agora se trata apenas um problema de encontrar os indivíduos que não são dominados

```python
TAMANHO_POPULACAO = 150
NUM_GERACOES = 150
CHANCE_DE_CRUZAMENTO = 0.5
CHANCE_DE_MUTACAO = 0.1
tamanho_torneio = 3
```
